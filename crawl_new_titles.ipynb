```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime
```

```python
def get_anue_titles_urls(url, pages:str, start:str = str(int(datetime.datetime(2023, 2, 19).timestamp())), end:str =str(int(datetime.datetime(2023, 2, 21).timestamp())-1)) :
    import requests as rq
    payload = {
    'limit':'30',
    'startAt': start,
    'endAt': end,
    'page': pages,
    'startAt': start,
    'endAt': end,
    'limit': '30'
    }
    response = rq.get(url, params=payload)
    news_urls = []
    news_titles = []
    for i in response.json()['items']['data']:
        news_urls += [i['newsId']]
        news_titles += [i['title']]
    return pd.DataFrame({'news_urls':news_urls, 'news_titles':news_titles})
    
```
